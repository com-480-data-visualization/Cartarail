```python
import pandas as pd
```

```python
stops = pd.read_csv("../data/gtfs_fp2025_2025-03-20/stops.txt")
stops
```

```python
stations = stops[stops["parent_station"].isna()]
stations = stations[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']].copy()
stations
```

```python
# Check: Every station's stop_id starts with 'Parent'.
assert(len(stations[~stations['stop_id'].str.contains("Parent")]) == 0)
```

```python
routes = pd.read_csv("../data/gtfs_fp2025_2025-03-20/routes.txt")
routes
```

```python
routes = routes[['route_id', 'route_short_name', 'route_long_name', 'route_desc', 'route_type']]
routes
```

```python
trips = pd.read_csv("../data/gtfs_fp2025_2025-03-20/trips.txt", dtype={'original_trip_id': str})
trips
```

```python
trips = trips[['route_id', 'trip_id']]
trips
```

```python
stop_times = pd.read_csv("../data/gtfs_fp2025_2025-03-20/stop_times.txt")
stop_times
```

```python
stop_times = stop_times.merge(stops[['stop_id', 'parent_station']], on='stop_id', how='left')
stop_times = stop_times[['trip_id', 'arrival_time', 'departure_time', 'parent_station', 'stop_sequence']]
stop_times
```

### Trains

```python
# Check https://opentransportdata.swiss/en/cookbook/gtfs/#routestxt for route types
train_routes = routes[(routes["route_type"] >= 100) & (routes["route_type"] < 200)]
train_routes
```

```python
train_trips = trips[trips["route_id"].isin(train_routes["route_id"])]
train_trips
```

```python
train_stop_times = stop_times[stop_times["trip_id"].isin(train_trips["trip_id"])]
train_stop_times
```

```python
train_stations = stations[stations['stop_id'].isin(train_stop_times['parent_station'])]
train_stations
```

```python
train_stations.to_csv("../data/preprocessed/train/stations.csv", index=False);
train_routes.to_csv("../data/preprocessed/train/routes.csv", index=False); 
train_trips.to_csv("../data/preprocessed/train/trips.csv", index=False);
train_stop_times.to_csv("../data/preprocessed/train/stop_times.csv", index=False);
```

### Lausanne

```python
lausanne_flon = stations[stations["stop_name"].str.contains("Lausanne-Flon")]
lausanne_flon_coords = (lausanne_flon.iloc[0]["stop_lat"], lausanne_flon.iloc[0]["stop_lon"])
lausanne_flon_coords
```

```python
from geopy.distance import geodesic

stations.loc[:, "distance_to_lausanne_flon"] = stations.apply(
    lambda row: geodesic(lausanne_flon_coords, (row["stop_lat"], row["stop_lon"])).km, axis=1
)
stations
```

```python
lausanne_stations = stations[stations["distance_to_lausanne_flon"] <= 10] 
lausanne_stations = lausanne_stations[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']]
lausanne_stations
```

```python
lausanne_stop_times = stop_times[stop_times['parent_station'].isin(lausanne_stations['stop_id'])]
lausanne_stop_times
```

```python
lausanne_trips = trips[trips['trip_id'].isin(lausanne_stop_times['trip_id'])]
lausanne_trips
```

```python
lausanne_routes = routes[routes['route_id'].isin(lausanne_trips['route_id'])]
lausanne_routes
```

```python
lausanne_stations.to_csv("../data/preprocessed/lausanne/stations.csv", index=False);
lausanne_trips.to_csv("../data/preprocessed/lausanne/trips.csv", index=False);
lausanne_routes.to_csv("../data/preprocessed/lausanne/routes.csv", index=False);
lausanne_stop_times.to_csv("../data/preprocessed/lausanne/stop_times.csv", index=False);
```
